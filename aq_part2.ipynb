```python
# eigenvectors: correlations between the time series of each precip record and the two principal components
sumv=0
sumc=0
#print('correlations between stn precipitation and first principal component time series')
for no in range(0,nos): 
#      print ("%s:  %.2f %.2f %.2f" % (stns[no+1],principal_axes[no,0],principal_axes[no,1],principal_axes[no,2]))
      sumv=sumv+principal_axes[no,0]**2
      sumc=sumc+principal_axes[no,0]*principal_axes[no,1]

print('check that the variance explained by the first principal component is the eigenvalue '\
      'and that spatially the first and second eigenvectors are uncorrelated')
print(" variance: %.1f correlation: %.1f"% ( sumv,sumc))

fig7,ax= plt.subplots(1,1,figsize=(15,7.5))
ax.scatter(principal_axes[:, 0],principal_axes[:, 1],marker='+')
for i in range(nos-1):
    ax.text(x=principal_axes[i, 0]+.004,y=principal_axes[i, 1]+.004,s=aqu.columns[i+1])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set(xlim=(.5,1))
ax.set(ylim=(-.65,.65))
ax.grid()
ax.title('U0035056 Correlations between Site Time Series and Principal Components 1 and 2')
plt.savefig('eigenvectors.png')
```

```python
#  principal component time series computed by multiplying the original time series by the weights
#  defined for each location and then scaled by the eigenvalues
principal_components = aqu_an.dot(principal_axes)/l

#create a dataframe of the principal components
pc = pd.DataFrame(data=principal_components)
basic_vals = pc.describe(percentiles=[.01,.50,.99])
#the following output is simply a check to see if the values of the time series seem appropriate
display(basic_vals)
```
